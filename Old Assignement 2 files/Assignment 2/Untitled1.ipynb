```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_theme()
import numpy as np
```

```python
df = pd.read_csv('avg_rocchio.csv')
```

```python
temp = df[df['beta'] == 0.15]
```

```python
temp = temp[temp['gamma'] == 0.2]
```

```python
temp
```

```python
X = temp['alpha']
Y = temp['nDCG']
plt.plot(X,Y)
plt.xlabel('Alpha')
plt.ylabel('nDCG score')
plt.title('nDCG score vs Alpha')
plt.savefig('nDCG_vs_alpha.png',dpi=1200)
plt.show()
```

```python
temp = df[df['alpha'] == 2.0]
temp = temp[temp['gamma'] == 0.2]
temp = temp[temp['beta']!=0.00]
temp
```

```python
X = temp['beta']
Y = temp['nDCG']
plt.plot(X,Y)
plt.xlabel('Beta')
plt.ylabel('nDCG score')
plt.title('nDCG score vs Beta')
plt.savefig('nDCG_vs_beta.png',dpi=1200)
plt.show()
```

```python
temp = df[df['alpha'] == 2.0]
temp = temp[temp['beta'] == 0.15]
temp = temp[temp['gamma']!=0.00]
temp
```

```python
X = temp['gamma']
Y = temp['nDCG']
plt.plot(X,Y)
plt.xlabel('Gamma')
plt.ylabel('nDCG score')
plt.title('nDCG score vs Gamma')
plt.savefig('nDCG_vs_gamma.png',dpi=1200)
plt.show()
```

```python
df = pd.read_csv('avg_LM.csv')
```

```python
df
```

```python
temp = df[:9]
temp
```

```python
X = np.log10(temp['mu'])
y1 = temp['nDCG_R1']
y2 = temp['nDCG_R2']
```

```python
plt.plot(X,y1,label='nDCG_R1')
plt.plot(X,y2,label='nDCG_R2')
plt.legend()
plt.xlabel('log10(mu)')
plt.ylabel('nDCG score')
plt.title('nDCG score vs mu')
plt.savefig('nDCG_vs_mu_log.png',dpi=1200)
plt.show()
```
